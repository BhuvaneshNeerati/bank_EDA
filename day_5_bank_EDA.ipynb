```python
## DATA ANALYSIS PROJECT STEPS : 
'''
# import data 
# exploratory data analysis(EDA)
# data preprocessing
# inferiential statistics(hypothesis testing) 
# feature engineering/Selection
# Model building
# model evaluation and comparision 
# model deployment

'''


''' 
Exploratory data analysis 
provides complete understanding of data(3 componetnts)
1) descriptive statistics
2) data visualization
3) data aggrigation
'''

'''
1)  descriptive statistics -  "Numeric variables" -univariate statistics

a) mean median mode 
b) variance and standard deviation
c) skewness and kurtosis(Normal distn)
d) quartiles , percentiles and deciles

-> NON NUMERIC VARIABLES - frequency counts
'''

```

```python
bank_data['balance'].describe() # give basic statistics

# if mean and median are not close , std_deviation very high -> "Indicates outlier and skewness"
```

```python
bank_data['age'].describe()
# here mean and median are close so, std is less (+/- 20%) -> indicates "no skewness"
```

```python
''' skew must be close to zero -> -ve rightside curve, +ve left side curve '''
print("skewness : ",bank_data['balance'].skew())
print("skewness : ",bank_data['age'].skew())
```

```python
'''kurtosis'''
# +ve (>3)-> tall and narrow
# -ve (<3)-> flat and wide peak
# (=3) -> normal distbution -> "hypothetical situation(basically never happens)"
print("kurtosis : ",bank_data['balance'].kurt())
print("kurtosis : ",bank_data['age'].kurt())
```

```python
'''
descriptiov statistics (non numeric )-> frequency counts
'''

print(bank_data['y'].value_counts())
print("----------------------------------------------")
print(bank_data['job'].value_counts())
```

```python
# EDA - data aggregation 
# groupby() and pivot_table()
```

```python
bank_data['age'].groupby(bank_data['y']).mean()

```

```python
# what is avg balance of differnet job categories?
bank_data['balance'].groupby(bank_data['job']).mean()
```

```python
# what is the median balance of different education?
bank_data['balance'].groupby(bank_data['education']).median()
```

```python
# what is the average balance of different y and education?
bank_data.pivot_table(index='education',columns='y',values='balance',aggfunc='mean')
```

```python
# what is mdeian balance of differnet job and education?
# bank_data.pivot_table(index='education',columns='job',values='balance',aggfunc='median')
# ( or ) -> keep leanthy one in index(rows) 

bank_data.pivot_table(index='job',columns='education',values='balance',aggfunc='median')


```

```python
bank_data.pivot_table(index='job',columns='education',values='balance',aggfunc='median').plot(kind='bar',stacked=True)
```

```python
# EDA - Non-numneric variables - cross tabulation- frequency counts of 2 non numnberic variables
pd.crosstab(bank_data['job'],bank_data['education'])
```
